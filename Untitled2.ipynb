```python
import stanza
```

```python
nlp_ru = stanza.Pipeline('ru')
```

```python
ACCEPTABLE_UPOS = ['NOUN']
def extract_nouns_from_text(text: str):
    nouns = []
    doc = nlp_ru(text)
    for sent in doc.sentences:
        for word in sent.words:
            if word.upos not in ACCEPTABLE_UPOS:
                    continue
            
            nouns.append(word.lemma)
    return nouns
```
