```python
def get_speciality_name_by_id(speciality_id: str):
    speciality_name = None
    if not speciality_id:
        return None
    
    items = specialty_df.loc[specialty_df['id'] == speciality_id]['name'].items()
    for item in items:
        if item[1] is not None:
            return item[1]
    
    return None
```

```python
df_h_ordered['speciality_name'] = df_h_ordered.apply(lambda row: get_speciality_name_by_id(row['specialty_id']), axis = 1)
```
