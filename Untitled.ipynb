```python
def get_speciality_name_by_id(speciality_id: str):
    speciality_name = None
    if not speciality_id:
        return None
    
    items = specialty_df.loc[specialty_df['id'] == speciality_id]['name'].items()
    for item in items:
        if item[1] is not None:
            return item[1]
    
    return None
```

```python
def find_symptom_id_list_by_disease_id(disease_id: int):
    symptom_id_list = ds_df.loc[ds_df['disease_id'] == disease_id]['symptom_id'].values
    return symptom_id_list
    
```

```python
def find_symptom_by_doctor_id(doctor_id: str):
    if not doctor_id:
        return []
    try:
        symptoms = []
        disease_id_list = find_disease_id_list_by_doctor_id(int(doctor_id))
        for disease_id in disease_id_list:
            symptom_id_list = find_symptom_id_list_by_disease_id(disease_id)
            cur_symptoms = symptom_df.loc[symptom_df['id'].isin(symptom_id_list)]['name'].values.tolist()
            symptoms += cur_symptoms
    except:
        return []
    
    return symptoms
```

```python
speciality_id_name = dict()
for idx, raw in specialty_df.iterrows():
    speciality_id_name[raw['id']] = raw['name']
```

```python
with open('./data/speciality_id_name_dict.json', 'w') as fp:
    json.dump(speciality_id_name, fp)
```
